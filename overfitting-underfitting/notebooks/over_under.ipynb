```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create simple dataset
np.random.seed(0)
X = np.linspace(-3, 3, 20)
y = 0.5 * X**2 + X + 2 + np.random.randn(20)  # Parabolic pattern with noise

# Step 2: Try different degrees one by one
degrees = [1, 2, 15]  # Underfit, good fit, overfit

for d in degrees:
    # Fit polynomial of degree d
    coeffs = np.polyfit(X, y, deg=d)
    model = np.poly1d(coeffs)

    # Predict on smooth curve
    X_test = np.linspace(-3, 3, 100)
    y_pred = model(X_test)

    # Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, color='blue', label='Data')
    plt.plot(X_test, y_pred, color='red', label=f'Degree {d} Fit')
    plt.title(f'Polynomial Fit: Degree {d}')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
```
