```python
import pandas as pd 
```

```python
# Load the dataset
df = pd.read_csv('../data/train.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style for better visual appearance
sns.set(style='whitegrid')

# 1. Visualize the number of survivors and non-survivors
sns.countplot(data=df, x='Survived')
plt.title('Survival Count (0 = Did Not Survive, 1 = Survived)')
plt.show()
```

```python
# Survival count by gender
# This shows how many men and women survived or not
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Count by Sex')
plt.show()

# Calculate actual survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

```python
# 3. Survival count by passenger class (Pclass)
# Visualizes the relation between ticket calss and survival
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Count by Passenger Class')
plt.show()

# Calculate survival rate by class 
df.groupby('Pclass')['Survived'].mean()
```

```python
# 4. Distribution of age
# This shows how age is distributed across all passengers
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Boxplot of Age vs Survival
# Compares the age of survivors vs non-survivors
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age vs Survival')
plt.show()
```

```python
# 5. Survival by Embarked port 
# This shows which boarding location (Embarked) had higher survival rates
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.title('Survival Count by Embarked Port')
plt.show()

# Calculate survival rate by port of embarkation
df.groupby('Embarked')['Survived'].mean()
```
